```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg', color='green')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
from scipy.stats import spearmanr
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, stat_func=spearmanr)
```

```python
sns.jointplot(x='size', y='total_bill', data=tips, stat_func=spearmanr)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, ratio=1, size=7, stat_func=spearmanr)
```
