```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mean = [0, 0]
cov =[[0.2, 0],[0,3]]
```

```python
x_axis, y_axis = np.random.multivariate_normal(mean, cov, size=40).T
```

```python
sns.kdeplot(x_axis)
```

```python
sns.kdeplot(y_axis)
```

```python
sns.kdeplot(x_axis, shade=True, color='green')
```

```python
sns.kdeplot(y_axis, color='blue', shade=True)
```

```python
sns.kdeplot(x_axis, y_axis)
```

```python
sns.kdeplot(x_axis, y_axis, shade=True, color='green')
```

```python
sns.kdeplot(x_axis, y_axis, cmap='coolwarm')
```

```python
sns.kdeplot(x_axis, y_axis, cmap='coolwarm', n_levels=35)
```

```python
sns.kdeplot(x_axis, bw=0.1)
```

```python
sns.kdeplot(y_axis, vertical=True)
```

```python
iris = sns.load_dataset('iris')
```

```python
setosa = iris.loc[iris.species == 'setosa']
versiclor = iris.loc[iris.species == 'versicolor']
```

```python
sns.kdeplot(setosa.petal_length, setosa.petal_width)
sns.kdeplot(versiclor.petal_length, versiclor.petal_width)
```

```python
sns.kdeplot(setosa.petal_length, setosa.petal_width, cmap='coolwarm')
sns.kdeplot(versiclor.petal_length, versiclor.petal_width, cmap='RdBu', shade=True)
```

```python
sns.kdeplot(setosa.petal_width, setosa.petal_length, cmap='coolwarm')
sns.kdeplot(versiclor.petal_width, versiclor.petal_length,  cmap='RdBu', n_levels=30)
```
