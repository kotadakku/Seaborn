```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.barplot(x='day', y='tip', data=tips)
```

```python
sns.barplot(x='day', y='total_bill',data=tips)
```

```python
sns.barplot(x='day', y='tip', data=tips, hue='sex')
```

```python
sns.barplot(x='day', y='tip', data=tips, hue='sex', palette ='winter_r')
```

```python
sns.barplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.barplot(x='total_bill', y='day', data=tips, palette='spring')
```

```python
sns.barplot(x='day', y='tip', data=tips, palette='spring', order=['Sat','Fri','Sun','Thur'])
```

```python
from numpy import median
sns.barplot(x='day',y='total_bill', data=tips, estimator=median, palette='spring')
```

```python
sns.barplot(x='smoker', y='tip', data=tips, estimator=median, hue='sex', palette='coolwarm')
```

```python
sns.barplot(x='smoker', y='tip', data=tips, ci=99)
```

```python
sns.barplot(x='smoker', y='tip', data=tips, ci=60)
```

```python
sns.barplot(x='smoker', y='tip', data=tips, ci=34, estimator=median, palette='winter_r' )
```
