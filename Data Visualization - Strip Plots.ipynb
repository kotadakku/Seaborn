```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
sns.stripplot(x=tips['tip'], color='green')
```

```python
sns.stripplot(x=tips['total_bill'])
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=False)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=1.5)
```

```python
sns.stripplot(x='total_bill', y='day', data=tips, jitter=1, linewidth=1.5)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='smoker', jitter=True, split=True, palette='winter_r')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='smoker', jitter=True, split=False, palette='winter_r')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='smoker', jitter=True, split=True, palette='winter_r', order=['Fri', 'Sun', 'Sat', 'Thur'])
```

```python
sns.stripplot(x='sex', y='tip', data=tips, order =['Female','Male'], palette='husl', marker='D')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, marker='D', size =15, edgecolor='gray', hue='sex', alpha=0.3, split=True, palette='husl')
```

```python
sns.stripplot(x='tip', y='day', data=tips,palette='winter_r')
sns.boxplot(x='tip', y='day', data=tips)
```

```python
sns.stripplot(x='tip', y='day', data=tips, jitter=True)
sns.violinplot(x='tip', y='day', data=tips, color='0.9')
```
