```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(iris, hue='species')
```

```python
sns.pairplot(iris, hue='species', palette='husl')
```

```python
sns.pairplot(iris, hue='species', palette='winter_r', markers=['o','D','s'])
```

```python
sns.pairplot(iris, hue='species', palette='husl', vars=['sepal_length','sepal_width'])
```

```python
sns.pairplot(iris, hue='species', palette='husl', vars=['sepal_length','sepal_width'], size=4)
```

```python
sns.pairplot(iris, x_vars=['petal_length','petal_width'], 
             y_vars=['sepal_length', 'sepal_width'],
            palette='husl', hue='species')
```

```python
sns.pairplot(iris, diag_kind='kde')
```

```python
sns.pairplot(iris, diag_kind='kde', palette='husl', hue='species')
```

```python
sns.pairplot(iris, diag_kind='kde', kind='reg', palette='husl', hue='species')
```
