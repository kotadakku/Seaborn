```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset('flights')
```

```python
flights = flights.pivot('month','year', 'passengers')
```

```python
sns.heatmap(flights)
```

```python
sns.clustermap(flights)
```

```python
sns.clustermap(flights, col_cluster=False)
```

```python
sns.clustermap(flights, row_cluster=False)
```

```python
sns.clustermap(flights, cmap='Blues_r', linewidths=1)
```

```python
sns.clustermap(flights, cmap='coolwarm', linewidths=1, figsize=(6,4))
```

```python
sns.clustermap(flights, standard_scale=1)
```

```python
sns.clustermap(flights, z_score=0)
```

```python
sns.clustermap(flights, z_score=1)
```
