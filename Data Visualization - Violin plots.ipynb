```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.violinplot(x=tips['tip'], y=tips['day'])
```

```python
sns.violinplot(x='size', y='total_bill', data = tips)
```

```python
sns.violinplot(x='day', y='total_bill', data = tips, palette='winter_r')
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data = tips, palette='husl')
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data = tips, palette='husl')
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette = 'coolwarm', order=['Sat', 'Thur', 'Sun', 'Fri'], hue='smoker', split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', palette='coolwarm', inner='quartile')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', 
               palette='RdBu', inner='quartile')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', 
               palette='RdBu', inner='quartile', split=True, scale='count')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', 
               palette='RdBu', inner='stick', split=True, scale='count')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', 
               palette='husl', inner='stick', scale='count')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', 
               palette='husl', inner='stick', scale='count', scale_hue=False)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', 
               palette='husl', inner='stick', scale='count', scale_hue=False
              ,split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', 
               palette='husl', inner='stick', scale='count', scale_hue=False
              ,split=True, bw=0.1)
```
