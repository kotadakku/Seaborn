```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
x= sns.PairGrid(iris)
x= x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(iris, hue='species')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(iris, hue='species', palette = 'coolwarm')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris, hue='species', palette = 'husl')
x = x.map_diag(plt.hist, histtype='step')
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris, hue='species', palette = 'husl')
x = x.map_diag(plt.hist, histtype='step', linewidth=4)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x= sns.PairGrid(iris, vars =['petal_length','petal_width'])
x= x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris, hue = 'species', vars =['petal_length','petal_width'])
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris, hue = 'species', vars =['petal_length','petal_width'])
x = x.map_diag(plt.hist, edgecolor='white')
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris, x_vars=['petal_length','petal_width'], y_vars=['sepal_length','sepal_width'])
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
```

```python
x = sns.PairGrid(iris, hue = 'species')
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris, palette='RdBu', hue = 'species', 
                 hue_kws ={'marker':['D','s','*']})
x = x.map(plt.scatter, s=50, edgecolor='black')
x = x.add_legend()
```
