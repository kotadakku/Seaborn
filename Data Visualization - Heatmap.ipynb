```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
normal = np.random.rand(12, 15)
```

```python
sns.heatmap(normal)
```

```python
sns.heatmap(normal, annot=True)
```

```python
sns.heatmap(normal, vmin=0, vmax =2)
```

```python
sns.heatmap(normal, vmin=0, vmax =3)
```

```python
sns.heatmap(normal, vmin=0, vmax =0.5)
```

```python
flights = sns.load_dataset('flights')
```

```python
flights = flights.pivot('month', 'year', 'passengers')
```

```python
sns.heatmap(flights)
```

```python
sns.heatmap(flights, annot=True)
```

```python
sns.heatmap(flights, annot=True, fmt='d')
```

```python
sns.heatmap(flights, annot=True, fmt='d', linewidths=1)
```

```python
sns.heatmap(flights, annot=True, fmt='d', linewidths=1, vmin=100, vmax=650)
```

```python
sns.heatmap(flights, cmap="summer", annot=True, fmt='d')
```

```python
sns.heatmap(flights, cmap="summer", annot=True, fmt='d', center= flights.loc['June',1954])
```

```python
sns.heatmap(flights, cmap="summer", annot=True, fmt='d', center= flights.loc['March',1959])
```

```python
sns.heatmap(flights, cmap="summer", annot=True, fmt='d', center= flights.loc['March',1959], cbar=False)
```

```python
sns.heatmap(flights, cbar=False, linewidths =1)
```
