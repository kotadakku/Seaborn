```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.FacetGrid(tips, row='smoker', col='time' )
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time' )
x = x.map(plt.hist, 'total_bill')
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time' )
x = x.map(plt.hist, 'total_bill', color='green', bins=15)
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time' )
x = x.map(plt.scatter, 'total_bill','tip')
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time' )
x = x.map(sns.regplot, 'total_bill', 'tip')
```

```python
x = sns.FacetGrid(tips, hue='smoker', col='time' )
x = x.map(plt.scatter, 'total_bill','tip') 
x = x.add_legend()
```

```python
x = sns.FacetGrid(tips, col='day')
x = x.map(sns.boxplot,'time', 'total_bill') 
x = x.add_legend()
```

```python
x = sns.FacetGrid(tips, col='day', size=7, aspect=.4)
x = x.map(sns.boxplot, 'total_bill','time') 
x = x.add_legend()
```

```python
x = sns.FacetGrid(tips, col='day', col_order=['Sat', 'Thur', 'Sun', 'Fri'], size=7, aspect=.4)
x = x.map(sns.boxplot, 'total_bill','time', color='red') 
x = x.add_legend()
```

```python
x = sns.FacetGrid(tips, hue='smoker', col='time', palette='husl', size=5, aspect=0.7 )
x = x.map(plt.scatter, 'total_bill','tip') 
x = x.add_legend()
```
