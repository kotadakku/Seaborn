```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
sns.regplot(x='total_bill', y='tip', data=tips, color='red')
```

```python
mean = [2,5]
cov = [[1.1,0.4],[2.2,3]]
x_value, y_value = np.random.multivariate_normal(mean, cov, 100).T
```

```python
sns.regplot(x=x_value, y= y_value)
```

```python
seriesx_value = pd.Series(x_value, name="Var-X")
seriesy_value = pd.Series(y_value, name="Var-Y")

sns.regplot(seriesx_value, seriesy_value)
```

```python
sns.regplot(seriesx_value, seriesy_value, marker='D')
```

```python
sns.regplot(x='total_bill', y='tip', data= tips, marker='D',  
            scatter_kws={'color':'red'}, 
            line_kws={'color':'red', 'linewidth':3.1})
```

```python
sns.regplot(x='total_bill', y='tip', data= tips, marker='*',  
            scatter_kws={'color':'gray'}, 
            line_kws={'color':'red', 'linewidth':3.1})
```

```python
sns.regplot(x=x_value, y=y_value, ci=70)
```

```python
sns.regplot(x='size', y='total_bill', data=tips)
```

```python
sns.regplot(x='size', y='total_bill', data=tips, x_jitter=0.6)
```

```python
sns.regplot(x='size', y='total_bill', data=tips, x_jitter=0.6, x_estimator=np.mean)
```
